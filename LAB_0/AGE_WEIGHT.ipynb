```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
x=np.array([5,6,7,8,9,10])
y=np.array([18,20,22,25,28,32])
plt.scatter(x,y,40,c='r',label='scatter_data',marker='x')
plt.xlabel('age')
plt.ylabel('weight')
plt.title('age_weight_graph')
plt.grid(True,color='k')
plt.show()
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
n=len(x)
```

```python
num=0
den=0
for i in range(n):
    num=num+((x[i]-mean_x)*(y[i]-mean_y))
    den=den+((x[i]-mean_x)**2)
m=num/den
c=mean_y-(m*mean_x)
print(m,c)
```

```python
y_pred=np.array([])
for i in range(n):
    p=(m*x[i]+c)
    y_pred=np.append(y_pred,np.array([p]),axis=0)
print(y_pred)
```

```python
plt.plot(x,y_pred,color='m',label='regression_line')
plt.scatter(x,y,40,c='g',label='actual_data',marker='o')
plt.scatter(x,y_pred,40,c='k',label='predicated data',marker='x')
plt.xlabel('population')
plt.ylabel('profit')
plt.legend()
plt.show()
```

```python
nu=0
de=0
for i in range(n):
    nu=nu+((y_pred[i]-mean_y)**2)
    de=de+((y[i]-mean_y)**2)
r=nu/de
r
```
