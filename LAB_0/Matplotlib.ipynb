# SAMPLE LINEAR GRAPH

```python
x1=np.array([1,3,4])
y1=np.array([2,4,6])
x2=np.array([6,4,3])
y2=np.array([5,2,1])
plt.plot(x1,y1,label='line one',color='c',linewidth=6)
plt.plot(x2,y2,label='line two',color='m',linewidth=6)
plt.title("sample linear graph")
plt.xlabel('plot number')
plt.ylabel('important var')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

# BAR GRAPH

```python
x1=np.array([5,8,10])
y1=np.array([12,6,6])
x2=np.array([6,9,10])
y2=np.array([6,15,7])
plt.bar(x1,y1,label='bar one',color='m')
plt.bar(x2,y2,label='bar two',color='c')
plt.legend()
plt.xlabel('bar number')
plt.ylabel('bar values')
plt.title('sample bar graph')
plt.show()
```

# HISTOGRAM

```python
x=[10,15,20,21,22,45,67,23,12,80,68]
y=[0,10,20,30,30,40,50,60,70,80]
plt.hist(x,y,histtype='bar',label='hist',rwidth=0.5)
plt.xlabel('x hist')
plt.ylabel('y hist')
plt.title('histogram')
plt.legend()
plt.show()
```

# SCATTER PLOT

```python
x1=np.array([5,3,10])
y1=np.array([12,6,6])
plt.scatter(x1,y1,label='scat',color='r')
plt.xlabel('x scat')
plt.ylabel('y scat')
plt.title('scatterplot')
plt.legend()
plt.show()
```

# stack/Area plot

```python
a=np.array([1,2,3,5,6,7])
b=np.array([2,3,4,6,7,8])
c=np.array([2,3,4,5,6,7])
d=np.array([8,5,6,8,1,2])
plt.plot([],[],color='g',label='green',linewidth=5)
plt.plot([],[],color='b',label='blue',linewidth=5)
plt.plot([],[],color='b',label='blue',linewidth=5)
plt.plot([],[],color='r',label='red',linewidth=5)
plt.stackplot(a,b,c,d,colors=['g','b','b','r'])
plt.xlabel('x scat')
plt.ylabel('y scat')
plt.title('stack plot')
plt.legend()
plt.show()
```

#  PIE CHART

```python
a=np.array([10,12,3,4])
b=np.array(['green','blue','red','black'])
colors=['g','b','r','k']
plt.pie(a,labels=b,colors=colors,startangle=90,shadow=True)
plt.title('piechart')
plt.show()
```
